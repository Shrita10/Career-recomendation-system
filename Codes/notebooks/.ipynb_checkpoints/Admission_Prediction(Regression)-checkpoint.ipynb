```python
df1.columns
df1 = df1.drop('Serial No.',axis = 1)
```

```python
X10 = df1.drop(['Chance of Admit'], axis=1)
y10 = df1['Chance of Admit']
```

```python
from sklearn.model_selection import train_test_split
X_train10, X_test10, y_train10, y_test10 = train_test_split(X10, y10, test_size = 0.2 , random_state = 50)
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(X_train10, y_train10)
```

```python
y_pred10 = regressor.predict(X_test10)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np
lr_r2 = r2_score(y_test10, y_pred10)
lr_mse = mean_squared_error(y_test10, y_pred10)
lr_rmse = np.sqrt(lr_mse)
print('SVR R2 Score: {0} \nSVR MSE: {1}, \nSVR RMSE:{2}'.format(lr_r2, lr_mse, lr_rmse))
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 500, random_state = 0)
regressor.fit(X_train10, y_train10)
```

```python
y_pred10 = regressor.predict(X_test10)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np
lr_r2 = r2_score(y_test10, y_pred10)
lr_mse = mean_squared_error(y_test10, y_pred10)
lr_rmse = np.sqrt(lr_mse)
print('Random forest R2 Score: {0} \nRandom Forest MSE: {1}, \nRandom Forest RMSE:{2}'.format(lr_r2, lr_mse, lr_rmse))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (9,7))
plt.scatter(y_test10, y_pred10)
plt.xlabel('Actual values')
plt.ylabel('Predicted values')
plt.title('Random Forest Regression', size = 15, weight = 'bold')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
lin_reg10 = LinearRegression()
lin_reg10.fit(X_train10, y_train10)
```

```python
y_pred10 = lin_reg10.predict(X_test10)
```

```python
pd.DataFrame({"Actual": y_test10, "Predict": y_pred10})
```

```python
from sklearn.metrics import r2_score, mean_squared_error
import numpy as np
lr_r2 = r2_score(y_test10, y_pred10)
lr_mse = mean_squared_error(y_test10, y_pred10)
lr_rmse = np.sqrt(lr_mse)
print('Linear Regression R2 Score: {0} \nLinear Regression MSE: {1}, \nLinear Regression RMSE:{2}'.format(lr_r2, lr_mse, lr_rmse))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (9,7))
plt.scatter(y_test10, y_pred10)
plt.xlabel('Actual values')
plt.ylabel('Predicted values')
plt.title('Linear Regression', size = 15, weight = 'bold')
plt.show()
```
